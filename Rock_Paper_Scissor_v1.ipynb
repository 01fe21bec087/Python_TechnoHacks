```python
import random
list1 = ["Rock","Paper","Scissors"]

def compare(computer, choice,compScore,userScore):
  if computer=="Rock":
    if choice=="Paper":
      print("You won!")
      userScore+=1
    elif choice=="Scissors":
      print("You lost")
      compScore+=1
    else:
      print("Draw, both chose ",choice)

  elif computer=="Paper":
    if choice=="Rock":
      print("You lost")
      compScore+=1
    elif choice=="Scissors":
      print("You won!")
      userScore+=1
    else:
      print("Draw, both chose ",choice)

  else:
#Computer=="Scissors"
    if choice=="Rock":
      print("You Won!")
      userScore+=1
    elif choice=="Paper":
      print("You lost")
      compScore+=1
    else:
      print("Draw, both chose ",choice)

  return compScore,userScore

compScore=0
userScore=0
while(1):
  choice = input("Choose Rock, Paper or Scissors (0 to exit): ")
  if (choice in list1):
    computer = random.choice(list1)
    print("Computer chose ",computer)
    compScore,userScore = compare(computer,choice,compScore,userScore)
    print("Scores: \n  User: ",userScore,"\n  Computer: ",compScore)
  else:
    print("Invalid character or Exit")
    break
```
